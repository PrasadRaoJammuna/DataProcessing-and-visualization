```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
views = np.array([123,932,567,789,345,893,563])
days = np.arange(1,8)
```

```python
plt.plot(days,views,label ='Youtube Views',color='g',marker='o',markerfacecolor='r')
plt.xlabel('Days')
plt.ylabel('Views')
plt.legend()
plt.title('Daily views for Youtube')
plt.show()
```

```python
plt.plot(days,views,label ='Youtube Views',color='g',
         marker='o',markerfacecolor='r',linestyle='--',
        linewidth=5)
plt.xlabel('Days')
plt.ylabel('Views')
plt.legend()
plt.title('Daily views for Youtube')
plt.show()
```

```python
y_views = np.array([932,567,789,345,893,563,876])
f_views = np.array([340,708,109,675,985,1020,580])
t_views = np.array([309,900,506,394,758,676,847])
days = np.arange(1,8)
```

```python
plt.plot(days,y_views,label='Youtube Views',marker='o')
plt.plot(days,f_views,label='Facebook Views',marker='*')
plt.plot(days,t_views,label='Twitter Views',marker='^')
plt.xlabel('Day')
plt.ylabel('Views')
plt.legend(loc=6)
plt.title('Daily views')
```

```python
plt.plot(days,y_views,label='Youtube Views',marker='o')
plt.plot(days,f_views,label='Facebook Views',marker='*')
plt.plot(days,t_views,label='Twitter Views',marker='^')
plt.xlabel('Day')
plt.ylabel('Views')
plt.xlim(1,7)
plt.ylim(100,900)
plt.legend(loc=6)
plt.title('Daily views')
```

```python
plt.plot(days,y_views,label='Youtube Views',marker='o')
plt.plot(days,f_views,label='Facebook Views',marker='*')
plt.plot(days,t_views,label='Twitter Views',marker='^')
plt.xlabel('Day')
plt.ylabel('Views')
plt.xlim(1,7)
plt.ylim(100,900)
plt.grid(True,linewidth=1.0,color='m',linestyle='-.')
plt.legend(loc=6)
plt.title('Daily views')
plt.savefig('views.png',dpi=500)
```

# Pie Charts

```python
labels = ['Facebook','Youtube','LinkedIn','Instagram']
views=[567,789,450,580]

plt.pie(views,labels=labels,explode=[0,0.2,0,0],autopct='%1.1f%%',
        shadow=True)
plt.show()
```

# Annote Text

```python
views = [503,932,567,789,345,893,563]
days = range(1,8)
```

```python
plt.plot(days,views,label ='Youtube Views',color='g',marker='o',markerfacecolor='r')

min_y = min(views)
x_coor = days[views.index(min_y)]

plt.xlabel('Days')
plt.ylabel('Views')
plt.legend()
plt.title('Daily views for Youtube')
plt.annotate('lowest value',xy=(x_coor,min_y),arrowprops=dict(facecolor='black',shrink= 0.7))
plt.show()
```
