```python
import pandas as pd
```

```python
df =pd.read_csv('climate.csv')
df
```

```python
#finding missing values in dataset
df.isnull().sum()
```

# Method 1

```python
new_df = df.fillna(
{
    'temparature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

# Method-2

```python
new_df=df.fillna(0) #it fill the values with zero
new_df
```

# method-ffill()/bfill()

```python
new_df = df.fillna(method='ffill')
new_df #it replaces the nan's with forward/previous values
```

```python
new_df = df.fillna(method='bfill')
new_df #it replaces the nan's with backward values
```

```python
new_df = df.fillna(method='ffill',limit=1)
new_df #it replaces the nan's with forward/previous values only once
```

## Drop na

```python
new_df = df.dropna()
new_df #drops all na rows
```

```python
new_df = df.dropna(how='all')
new_df #if row contains all na values then only it will drop
```

```python
new_df = df.dropna(thresh=3)
new_df #it limits the na values in rows and drop those rows
```
