# Like SQL

```python
# finding temparature greater than 35

df[df['Temperature']>35]
```

```python
#finding humidity when temparature >35

df[df['Temperature']>35][['Humidity','Temperature']]
```

```python
# Find windspeed when temparature max and min

df[df['Temperature']==df['Temperature'].min()][['Temperature','WindSpeedMPH']]
```

```python

df[df['Temperature']==df['Temperature'].max()][['Temperature','WindSpeedMPH']]
```
